```python
import os
import csv
import time
from lxml import html
import requests
from concurrent.futures import ThreadPoolExecutor, as_completed

# Function to fetch page content using a proxy with retry logic
def fetch_page(url, retries=3):
    proxy_url = 'https://proxy.scrapeops.io/v1/'
    api_key = 'de1f98bd-2f41-45b8-bef1-3228973a8243'

    for attempt in range(retries):
        try:
            response = requests.get(
                proxy_url,
                params={
                    'api_key': api_key,
                    'url': url,
                },
            )
            response.raise_for_status()
            return response.content
        except requests.RequestException as e:
            print(f"Request failed: {e}")
            if attempt < retries - 1:
                time.sleep(2 ** attempt)  # Exponential backoff
            else:
                return None

# Function to extract the next page URL
def get_next_page_url(tree):
    next_page_element = tree.xpath("//a[@aria-label='Next page']")
    if next_page_element:
        next_page_url = next_page_element[0].get('href')
        if next_page_url:
            return base_url + next_page_url
    return None

# Updated extract_data function with error handling
def extract_data(page_content):
    try:
        tree = html.fromstring(page_content)

        # Applying provided CSS selector
        name_elements = tree.cssselect(".biGQs._P.egaXP.rRtyp")
        a_elements = tree.xpath("//a[@aria-label='Email']")
        elements = tree.xpath("//a[@aria-label='Call']")
        add_link = tree.xpath("//div[@class='hpxwy e j']/a[contains(@href, 'maps')]")
        add_text = tree.xpath("//div[@class='hpxwy e j']/a[contains(@href, 'maps')]/span")
        cuisine_elements = tree.xpath("//div[@class='biGQs _P ncFvv NaqPn'][normalize-space()='CUISINES']/parent::div/following-sibling::div[@class='biGQs _P pZUbB alXOW eWlDX GzNcM ATzgx UTQMg TwpTY hmDzD']")

        name = [name.text_content().strip() for name in name_elements if name.text_content().strip()]
        cuisines = [div.text_content().strip() for div in cuisine_elements if div.text_content().strip()]

        telephone = [a.get('href') for a in elements if a.get('href')]
        email = [a.get('href') for a in a_elements if a.get('href')]
        text_address = [span.text_content().strip() for span in add_text if span.text_content().strip()]
        map_link = [a.get('href') for a in add_link if a.get('href')]

        return {
            'name': name,
            'telephone': telephone,
            'email': email,
            'map_link': map_link,
            'text_address': text_address,
            'cuisines': cuisines
        }
    except Exception as e:
        print(f"Error occurred in extract_data: {e}")
        return {
            'name': None,
            'telephone': None,
            'email': None,
            'map_link': None,
            'text_address': None,
            'cuisines': None
        }

# Base URL
base_url = 'https://www.tripadvisor.co.uk'
# Fetch the main page content
main_page_url = 'https://www.tripadvisor.co.uk/Restaurants-g186338-oa0-London_England.html'
current_page_url = main_page_url

# List to store the extracted data
data = []

# Function to handle the extraction of a single restaurant page
def handle_restaurant_page(href):
    page_content = fetch_page(href)
    if page_content:
        return extract_data(page_content)
    return None

# Function to process a single main page
def process_main_page(page_url):
    global current_page_url
    page_content = fetch_page(page_url)
    if not page_content:
        return None, None

    tree = html.fromstring(page_content)
    div_elements = tree.xpath("//div[@class='biGQs _P fiohW alXOW NwcxK GzNcM ytVPx UTQMg RnEEZ ngXxk']")
    hrefs = [base_url + anchor.get('href') for div in div_elements for anchor in div.xpath(".//a[@href]")]

    next_page_url = get_next_page_url(tree)
    return hrefs, next_page_url

# Loop through the next 110 pages
for _ in range(50):
    hrefs, next_page_url = process_main_page(current_page_url)
    if not hrefs:
        break

    for href in hrefs:
        result = handle_restaurant_page(href)
        if result and result['name']:
            data.append(result)
            print(result)

        # Adding a delay to avoid hitting the rate limit
        time.sleep(1)

    # If there is no next page, break the loop
    if not next_page_url:
        break

    current_page_url = next_page_url
    print(f"Current page URL: {current_page_url}")
    print(f"Next page URL: {next_page_url}")

# Define the CSV file path
csv_file_path = os.path.join(os.getcwd(), 'data_extracted_2.csv')

# Save the data to a CSV file
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['name', 'telephone', 'email', 'map_link', 'text_address', 'cuisines']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for item in data:
        writer.writerow(item)

print(f'Data has been saved to {csv_file_path}')
```
